By looking at the heatmap, we can identify which variables are correlated with each other. in this heatmap, we can see that the target variable median_house_value has negative correlation of -0.48 with the ocean_proximity_INLAND which is a moderete negative relationship, which indicates that if we are looking for picking a high value house, houses that are further from the ocean are not the way to go.

Also a somewhat strong correlation of 0.69 between median_house_value to median_income indicates that if we are looking to pick a house that is going to be valuable, it is usefull to look at the median income of people living nearby.
